```python
import pandas as pd

data = {'col1': [1, 2, 3], 'col2': ['A', 'B', 'C']}
df = pd.DataFrame(data)

new_row = {'col1': 4, 'col2': 'D'}
df.loc[4] = new_row

print(df)
```

```python
import pandas as pd

data = {'col1': [1, 2, 3], 'col2': ['A', 'B', 'C']}
df = pd.DataFrame(data)

new_row = {'col1': 5, 'col2': 'E'}
df.iloc[2] = new_row

print(df)
```

```python
import pandas as pd

data = {'col1': [1, 2, 3, 4], 'col2': ['A', 'B', 'C', 'D']}
df = pd.DataFrame(data)

# Drop the row at index 2
df = df.drop(2)

print(df)
```

```python
import pandas as pd

data = {'col1': [1, 2, 3], 'col2': ['A', 'B', 'C']}
df = pd.DataFrame(data, index=['a', 'b', 'c'])  # Set custom index

new_row = {'col1': 4, 'col2': 'D'}
df.loc['d'] = new_row

print(df)
```

```python
import pandas as pd

data = {'col1': [1, 2, 3], 'col2': ['A', 'B', 'C']}
df = pd.DataFrame(data)

new_row = {'col1': 5, 'col2': 'E'}
df.iloc[2] = new_row  # Insert at index 2, shifting existing row at 2 downwards

print(df)
```
