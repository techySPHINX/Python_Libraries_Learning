```python
#use annotation

import matplotlib.pyplot as plt

x = ["Python", "C", "C++", "Java"]
y = [85, 70, 60, 82]
c = ["y", "b", "m", "g"]


bars = plt.bar(x, y, width=0.3, color=c, edgecolor="r", label="Jagan")
plt.xlim(-0.2, 3.8)
plt.ylim(0, max(y) + 10)

# plt.figure(figsize=(10, 6))  # Set larger figure size (adjust as needed)
# plt.subplots_adjust(top=0.8)


for bar, value in zip(bars, y):
  plt.annotate(str(value),  # Text to display
               xy=(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.3),  # Position (center above bar with some offset)
               ha='center',
               va='bottom',
               fontsize=12,
               weight='bold')

plt.xlabel("Language", fontsize=15)
plt.ylabel("No", fontsize=15)
plt.title("Comparison of Programming Languages", fontsize=16)
plt.xticks(rotation=0, ha="center")
plt.ylabel("No", fontsize=15, ha="center")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["python","c","c++","Java"]
y=[85,70,60,82]
z=[20,30,40,50]
plt.xlabel("language",fontsize=15)
plt.ylabel("No",fontsize=15)
width = 0.2
plt.bar(x,y,width,color="y",label="Popularity")
plt.bar(x,z,width,color="r",label="Popularity1")

plt.ylabel("No", fontsize=15, ha="center")
plt.title("Domain",fontsize=20)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=["python","c","c++","Java"]
y=[85,70,60,82]
z=[20,30,40,50]

width = 0.2
p= np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("No",fontsize=15)

plt.bar(p,y,width,color="y",label="Popularity")
plt.bar(p1,z,width,color="r",label="Popularity1")

plt.xticks(p+width/2,x,rotation=90)
plt.title("Domain",fontsize=20)
plt.tight_layout()
plt.legend()
plt.show()
```

HoriZontal Bar Graph


```python
import matplotlib.pyplot as plt
import numpy as np

y=["python","c","c++","Java"]
x=[85,70,60,82]
z=[20,30,40,50]

width = 0.2
p= np.arange(len(y))
p1 = [j+width for j in p]

plt.ylabel("Language",fontsize=15)
plt.xlabel("No",fontsize=15)

plt.barh(p,x,width,color="y",label="Popularity")
plt.barh(p1,z,width,color="r",label="Popularity1")

plt.yticks(p+width/2,y)
plt.title("Domain",fontsize=20)
plt.tight_layout()
plt.legend()
plt.show()
```
